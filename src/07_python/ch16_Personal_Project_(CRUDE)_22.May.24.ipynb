```python
# api를 통해 획득한 데이터를 딕셔너리 형태로 저장하고 이를 dictwrite로 csv파일을 만들어보자.

import requests
import pandas as pd
from bs4 import BeautifulSoup

resp_attr = requests.get('http://openAPI.seoul.go.kr:8088/4a6353746b796a68333079486b464e/xml/SearchCulturalFacilitiesAddressService/1/1000')
soup = BeautifulSoup(resp_attr.content, 'xml')
            # 서버로부터 전송받은 xml데이터를 파일로 생성하는 단계를 추가할지 고민해보자.
attr_list = []
for row in soup.select('row'):
    fac_code = row.find('FAC_CODE').text
    subjcode = row.find('SUBJCODE').text
    codename = row.find('CODENAME').text
    fac_name = row.find('FAC_NAME').text
    addr = row.find('ADDR').text
#     attr_list.append( (fac_code, subjcode, codename, fac_name) )
    attr_list.append( {'fac_code' : fac_code, 'subjcode' : subjcode, 'codename' : codename, 'fac_name' : fac_name, 'addr' : addr} )
df_attr = pd.DataFrame(attr_list, columns = ['fac_code','subjcode', 'codename', 'fac_name', 'addr'])

import csv
url = 'http://api.vworld.kr/req/address?service=address&request=getCoord&version=2.0&crs=epsg:4326&format=xml&key=EA773332-297D-38CA-9A54-76F66D21A455&type=parcel'
dic_list = [] # 바로 위에서는 [리스트]를 생성해 데이터를 관리했는데, 여기서는 {딕셔너리}로 만들어 봄.
for idx, addr in enumerate(df_attr['addr']):
    # 확인할 때 이부분에 idx <= 100과 같이 빠른 실행을 위한 조건부여
    try:
        resp = requests.get(url + '&address=' + addr)
        soup = BeautifulSoup(resp.content, 'xml')
        if soup.select_one('status').text == 'OK':   # select 매서드를 사용하면 대상이 하나뿐인 경우에도 결국 에러가 발생.
            dic = {}
            dic['dist'] = soup.select_one('level2').text
            dic['x'] = soup.select_one('point > x').text
            dic['y'] = soup.select_one('point > y').text
            dic_list.append(dic)
        elif soup.select_one('status').text == 'NOT_FOUND': # ROAD형식의 주소를 PARCEL형식으로 요청했을 경우 NOT_FOUND가 반환되길래.
            url = 'http://api.vworld.kr/req/address?service=address&request=getCoord&version=2.0&crs=epsg:4326&format=xml&key=EA773332-297D-38CA-9A54-76F66D21A455&type=road'
            resp = requests.get(url + '&address' + addr)
            soup = BeautifulSoup(resp.content, 'xml')
            if soup.select_one('status').text == 'OK':
                dic = {}
                dic['dist'] = soup.select_one('level2').text
                dic['x'] = soup.select_one('point > x').text
                dic['y'] = soup.select_one('point > y').text
                dic_list.append(dic)
            else:
                dic = {}
                dic['dist'] = None
                dic['x'] = None
                dic['y'] = None
                dic_list.append(dic)
        else:
            dic = {}
            dic['dist'] = None
            dic['x'] = None
            dic['y'] = None
            dic_list.append(dic)
    except Exception :
        dic = {}
        dic['dist'] = None
        dic['x'] = None
        dic['y'] = None
        dic_list.append(dic)
df_coord = pd.DataFrame(dic_list, columns=['dist', 'x', 'y']) 

# for문을 통해 얻어진 데이터영역(district, x, y)을 딕셔너리 형태로 구성한 뒤 3 columns만 단독으로 일단 CSV파일을 생성해봄.
with open('ch15_distxy_dic.csv', 'w', encoding='cp949',newline='') as f:
    fieldnames = ['dist','x','y']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(dic_list)
            # 여기서는 이전 단계의 데이터와 함께 하나의 데이터프레임으로 만들어 최종 하나의 CSV파일을 만들어야 하므로
            # csv.Dictwriter를 사용하기보다 개별 데이터프레임으로 완성한 후 
            # 앞 단계의 데이터 또한 데이터프레임의 형태로 통일시킨 뒤 concat([df_a, df_b]. axis=1)으로 합치는게 좋을듯.

# diclist 형태로 가지고 있으면 1) 데이터프레임으로 형태변화 2) 파일로 쓰기 두 기능을 모두 적용시킬 수 있음.
# 리스트들에 pd.DataFrame을 적용시키기 전에 합침.
# temp = list(zip(dist_list, xcoord_list, ycoord_list))
# temp
```
